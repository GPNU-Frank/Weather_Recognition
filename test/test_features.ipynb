```python
# 测试分块
img_cloudy = cv2.resize(img_cloudy, (224, 244))

row, col = 2, 4
img_patch = [[None] * 4 for _ in range(2)]

for i in range(row):
    for j in range(col):
        img_patch[i][j] = img_cloudy[224 // row * i: 224 // row * (i + 1), 224 // col * j: 224 // col * (j + 1), :]
 
print(img_patch[0][1].shape)
# cv2.imshow('1', img_patch[1][3])
# k = cv2.waitKey(0)
# cv2.destroyAllWindows()
# plt.imshow(img_patch[1][0])
# plt.show()
img_contrast = [[None] * 4 for _ in range(2)]
for i in range(row):
    for j in range(col):
        gray_img = cv2.cvtColor(img_patch[i][j], cv2.COLOR_BGR2GRAY)
        img_contrast[i][j] = (gray_img.max() - gray_img.min()) / (gray_img.max() + gray_img.min())

print(img_contrast)
```
