```python
# 可视化 对比度 饱和度等特征
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_img_cloudy = "G:\\weather_recognition\\groud_truth\\cloudy\\cloud113.jpg"
path_img_sunny = "G:\\weather_recognition\\groud_truth\\sunny\\sun379.jpg"
path_img_rain = "G:\\weather_recognition\\groud_truth\\rainy\\rain247.jpg"
path_img_fog = "G:\\weather_recognition\\groud_truth\\haze\\fog1761.jpg"
```

```python
img_cloudy = cv2.imread(path_img_cloudy)

cv2.namedWindow("1", cv2.WINDOW_AUTOSIZE)
cv2.imshow("1", img_cloudy)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_cloudy_hsv = cv2.cvtColor(img_cloudy, cv2.COLOR_BGR2HSV)
s_hist = cv2.calcHist([img_cloudy_hsv], [1], None, [256], [0.0, 255.0])
plt.plot(s_hist)
plt.show()
```

```python
img_sunny = cv2.imread(path_img_sunny)
img_sunny_hsv = cv2.cvtColor(img_sunny, cv2.COLOR_BGR2HSV)
sunny_s_hist = cv2.calcHist([img_sunny_hsv], [1], None, [256], [0.0, 255.0])
plt.plot(sunny_s_hist)
plt.show()
```

```python
img_rain = cv2.imread(path_img_rain)
img_rain_hsv = cv2.cvtColor(img_rain, cv2.COLOR_BGR2HSV)
rain_s_hist = cv2.calcHist([img_rain_hsv], [1], None, [256], [0.0, 255.0])
plt.plot(rain_s_hist)
plt.show()
for idx, val in enumerate(rain_s_hist):
    print(idx, val)
```

```python
img_fog = cv2.imread(path_img_fog)
img_fog_hsv = cv2.cvtColor(img_fog, cv2.COLOR_BGR2HSV)
fog_s_hist = cv2.calcHist([img_fog_hsv], [1], None, [256], [0.0, 255.0])
plt.plot(fog_s_hist)
plt.show()
for idx, val in enumerate(fog_s_hist):
    print(idx, val)
```

```python
# canny 算子
img1 = cv2.Canny(img_cloudy, 100, 150)
cv2.imshow('1', img1)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 测试分块
img_cloudy = cv2.resize(img_cloudy, (224, 244))

row, col = 2, 4
img_patch = [[None] * 4 for _ in range(2)]

for i in range(row):
    for j in range(col):
        img_patch[i][j] = img_cloudy[224 // row * i: 224 // row * (i + 1), 224 // col * j: 224 // col * (j + 1), :]
 
print(img_patch[0][1].shape)
# cv2.imshow('1', img_patch[1][3])
# k = cv2.waitKey(0)
# cv2.destroyAllWindows()
# plt.imshow(img_patch[1][0])
# plt.show()
img_contrast = [[None] * 4 for _ in range(2)]
for i in range(row):
    for j in range(col):
        gray_img = cv2.cvtColor(img_patch[i][j], cv2.COLOR_BGR2GRAY)
        img_contrast[i][j] = (gray_img.max() - gray_img.min()) / (gray_img.max() + gray_img.min())

print(img_contrast)
```
