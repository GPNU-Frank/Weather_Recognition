```python
# 可视化 对比度 饱和度等特征
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 真实数据集
# "G:\\vscode_workspace\\Weather_Recognition\\data_split_v2\\train\\Cloud"
path_img_cloudy = "G:\\vscode_workspace\\Weather_Recognition\\data_split_v6\\train\\Cloud\\10548_2021_01_25 15_19_59_31_0.jpg"
path_img_sunny = "G:\\vscode_workspace\\Weather_Recognition\\data_split_v6\\test\\Sunny\\13292_2021_01_13 11_17_05_32_0.jpg"
path_img_rain = "G:\\weather_recognition\\groud_truth\\rainy\\rain247.jpg"
path_img_fog = "G:\\vscode_workspace\\Weather_Recognition\\data_split_v2\\train\\Fog\\3192_8348_2020_12_30 13_01_07_8.jpg"
```

```python
img_sunny = cv2.imread(path_img_sunny)

cv2.namedWindow("1", cv2.WINDOW_AUTOSIZE)
cv2.imshow("1", img_sunny)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_sunny = cv2.imread(path_img_sunny)
img_sunny_lab = cv2.cvtColor(img_sunny, cv2.COLOR_RGB2LAB)
sunny_l_hist = cv2.calcHist([img_sunny_lab], [0], None, [256], [0.0, 255.0])
plt.plot(sunny_l_hist)
plt.show()
```

```python
img_sunny_hsv = cv2.cvtColor(img_sunny, cv2.COLOR_RGB2HSV)
sunny_s_hist = cv2.calcHist([img_sunny_hsv], [1], None, [256], [0.0, 255.0])
plt.plot(sunny_s_hist)
plt.show()
```

```python
img_cloudy = cv2.imread(path_img_cloudy)
img_cloudy_lab = cv2.cvtColor(img_cloudy, cv2.COLOR_RGB2LAB)
cloudy_l_hist = cv2.calcHist([img_cloudy_lab], [0], None, [256], [0.0, 255.0])
plt.plot(cloudy_l_hist)
plt.show()
```
